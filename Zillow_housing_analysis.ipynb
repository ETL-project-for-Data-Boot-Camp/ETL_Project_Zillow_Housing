```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#load file
file_housing = "Resources/Zillow Single Family Residence.csv"
file_lat ="Resources/us-zip-code-latitude-and-longitude.csv"
```

```python
#create DataFrame from csv
df_housing = pd.read_csv(file_housing)
df_lat = pd.read_csv(file_lat)
df_lat
```

```python
#Simplified df_housing
df_housing = df_housing[[
 'RegionName',
 'City',
 'State',
 'CountyName',
 '2018-01',
 '2018-02',
 '2018-03',
 '2018-04',
 '2018-05',
 '2018-06',
 '2018-07',
 '2018-08',
 '2018-09',
 '2018-10',
 '2018-11',
 '2018-12']]
df_housing
```

```python
#simplied Lat file
df_lat = df_lat[['Zip','Latitude','Longitude']]
```

```python
#join the data_
merge_file = df_housing.merge(df_lat,left_on="RegionName",right_on = "Zip",how="left")
```

```python
merge_file = merge_file [['Zip',
 'City',
 'State',
'Latitude',
 'Longitude',
 'CountyName',
 '2018-01',
 '2018-02',
 '2018-03',
 '2018-04',
 '2018-05',
 '2018-06',
 '2018-07',
 '2018-08',
 '2018-09',
 '2018-10',
 '2018-11',
 '2018-12',
 ]]
```

```python
merge_file
```

```python
#merge 2 dataframes
merge_file1 = merge_file.dropna()
merge_file2= merge_file1.astype({'Zip': 'int'})

#rename dataframe to match sql format
df_zillow= merge_file2.rename(columns={"Zip": "zip", "City": "city","State":'state','Latitude':'latitude','Longitude':'longitude',
                         'CountyName':'countyname'})
df_zillow.head()
```

```python
#Connect to local Database
rds_connection_string = "postgres:246800@localhost:5432/zillow_housing"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#check for tables
engine.table_names()
```

```python
#use pandas to load csv converted DataFrame into database
df_zillow.to_sql(name='zillow',con=engine,if_exists='append',index=False)
```

```python
#Confirm data has been added by querying the customer_name table
pd.read_sql_query('select * from zillow',con=engine).head()
```
